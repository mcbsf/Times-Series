### b)

```python
mean = serie.mean() 
print(mean)
```

### c)

```python
def autocovariation(serie, lag):
    parcial = len(serie)**-1
    somatory = 0
    rangeMax = len(serie)-lag-1
    for t in range(0, rangeMax):
        somatory = somatory + (serie[t+lag] - mean)*(serie[t] - mean)
    return somatory*parcial
```

### d)

```python
def autocorrelation(serie, lag):
    return autocovariation(serie, lag)/autocovariation(serie, 0)
```

### b)

```python
print(ts.adfuller(serie_colorado, 1))
print(ts.adfuller(serie_eletricity, 1))
print(ts.adfuller(serie_lynx, 1))
print(ts.adfuller(serie, 1))
```

## 4ª

resposta: nenhumas das 3 são estacionárias. a primeira forte evidência é que o ADF retorna valores que divergem de 0, além de que as médias das séries não beiram 0, a outra evidência é que na função de correlação total é possível detectar tendencia e sazonalidade
